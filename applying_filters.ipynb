```python
import cv2
import numpy as np
import matplotlib.pyplot as plt


image_path = "tiger.jpg"  
image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)


plt.imshow(image, cmap="gray")
plt.title("Original Image")
plt.axis("off")
plt.show()
```

```python

sobel_x = cv2.Sobel(image, cv2.CV_64F, 1, 0, ksize=3)  
sobel_y = cv2.Sobel(image, cv2.CV_64F, 0, 1, ksize=3)  
sobel_combined = cv2.magnitude(sobel_x, sobel_y)  

def normalize(img):
    return cv2.convertScaleAbs(img)


sobel_x = normalize(sobel_x)
sobel_y = normalize(sobel_y)
sobel_combined = normalize(sobel_combined)


fig, axes = plt.subplots(1, 3, figsize=(12, 4))
axes[0].imshow(sobel_x, cmap="gray")
axes[0].set_title("Sobel X")
axes[1].imshow(sobel_y, cmap="gray")
axes[1].set_title("Sobel Y")
axes[2].imshow(sobel_combined, cmap="gray")
axes[2].set_title("Sobel Combined")

plt.tight_layout()
plt.show()
```

```python

scharr_x = cv2.Scharr(image, cv2.CV_64F, 1, 0)  
scharr_y = cv2.Scharr(image, cv2.CV_64F, 0, 1)  
scharr_combined = cv2.magnitude(scharr_x, scharr_y) 


scharr_x = normalize(scharr_x)
scharr_y = normalize(scharr_y)
scharr_combined = normalize(scharr_combined)


fig, axes = plt.subplots(1, 3, figsize=(12, 4))
axes[0].imshow(scharr_x, cmap="gray")
axes[0].set_title("Scharr X")
axes[1].imshow(scharr_y, cmap="gray")
axes[1].set_title("Scharr Y")
axes[2].imshow(scharr_combined, cmap="gray")
axes[2].set_title("Scharr Combined")

plt.tight_layout()
plt.show()
```

```python

prewitt_x = np.array([[-1, 0, 1], [-1, 0, 1], [-1, 0, 1]], dtype=np.float32) 
prewitt_y = np.array([[-1, -1, -1], [0, 0, 0], [1, 1, 1]], dtype=np.float32)  


vertical_edges = cv2.filter2D(image, cv2.CV_64F, prewitt_x)
horizontal_edges = cv2.filter2D(image, cv2.CV_64F, prewitt_y)
prewitt_combined = cv2.magnitude(vertical_edges, horizontal_edges)


vertical_edges = normalize(vertical_edges)
horizontal_edges = normalize(horizontal_edges)
prewitt_combined = normalize(prewitt_combined)


fig, axes = plt.subplots(1, 3, figsize=(12, 4))
axes[0].imshow(vertical_edges, cmap="gray")
axes[0].set_title("Vertical Edges (Prewitt)")
axes[1].imshow(horizontal_edges, cmap="gray")
axes[1].set_title("Horizontal Edges (Prewitt)")
axes[2].imshow(prewitt_combined, cmap="gray")
axes[2].set_title("Prewitt Combined")

plt.tight_layout()
plt.show()
```
